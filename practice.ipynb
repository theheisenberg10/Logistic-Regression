```python
def age1(col) :
    Age = col[0]
    Pclass = col[1]
    
    if pd.isnull(Age):
        
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 30
        else :
            return 24
    else :
        return Age
    
```

```python
train['Age']=train[['Age','Pclass']].apply(age1,axis=1)
```

```python
train
```

```python
train.drop('Cabin',axis=1,inplace=True)
```

```python
train.head()
```

```python
train.drop('Ticket',axis=1,inplace=True)
```

```python
male=pd.get_dummies(train['Sex'],drop_first=True)
embark = pd.get_dummies(train['Embarked'],drop_first=True)
```

```python
train=pd.concat([train,male,embark],axis=1)
```
